```python
import os
import shutil
import pandas as pd
import time
import glob
```

```python
# mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# path to Google Drive
google_drive_path = '/content/drive/My Drive/cs109b_final_project/'
```

```python
meta_filepath = google_drive_path + 'meta_data/' + 'Sample_Normal.csv'

image_paths_all = list()

infile_subfolders = ["images_001", "images_002", "images_003", "images_004", "images_005", "images_006", "images_007", "images_008", "images_009", "images_010", "images_011", "images_012"]
for infile_subfolder in infile_subfolders:
  infile_subfolder_path = google_drive_path + 'images_downsized/' + infile_subfolder + '/' + '*.png'
  image_paths_subfolder = list(glob.iglob(infile_subfolder_path))
  image_paths_all.append(image_paths_subfolder)

print(image_paths_all[:3])

image_paths_all_flatten = [path for sublist in image_paths_all for path in sublist]

image_names = list()
for image_path in image_paths_all_flatten:
  image_name = image_path.split('/')[-1]
  image_names.append(image_name)

print(image_names[:3])

meta_df = pd.read_csv(meta_filepath)

counter = 0

problem_names = list()

start_time = time.time()
for i in range(len(meta_df)):
  print('i={}'.format(i))
  image_name = meta_df['Image Index'][i]
  if image_name in image_names:
    matched_index = image_names.index(image_name)
    src = image_paths_all_flatten[matched_index]
    dst = google_drive_path + 'images_rescaled_subsample/' + image_name
    shutil.copyfile(src,dst)
    counter = counter + 1
  else:
    problem_names.append(image_name)

print("success counter={}".format(counter))
print(problem_names)

execution_time = (time.time() - start_time)/60.0
print("Training execution time (mins)",execution_time)
```
